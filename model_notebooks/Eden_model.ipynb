```
import numpy as np
import matplotlib.pyplot as plt
```

```
def Eden():
    t = 5000
    size = 128
    h = np.zeros((size,size))
    m = int(size/2)
    r = int(size/20)
    for i in range(size):
        for j in range(size):
            if (i-m)**2 + (j-m)**2 <= r**2 :
                h[i,j]=1

    for i in range(5):
        noiseMatrix = np.random.normal(5,1,(size,size))
        neighborBlackMatrix = np.sign((np.roll(h,1,axis=0)+
                                np.roll(h,-1,axis=0)+
                                np.roll(h,1,axis=1)+
                                np.roll(h,-1,axis=1))*(1-h))

        nextPosition = np.max((noiseMatrix * neighborBlackMatrix))
        MAT = (noiseMatrix * neighborBlackMatrix)/nextPosition
        floorArray = np.floor(MAT)
        BoolArray = np.array(floorArray == 1
        nextGrowthMatrix = BoolArray.astype(np.int)
        h = h + nextGrowthMatrix
    return h
```

```
h = Eden()
plt.imshow(t)
plt.show()
```
