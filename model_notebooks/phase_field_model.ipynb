```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import *
```

```
def phase_field(alpha,du,dv,thresh=1):
    dx = 0.2
    img_size = 128
    dt = 0.05

    def u_func(u, v):
        grad = u*(1-u)*(u-0.5+alpha*(v-0.5))
        return u + dt*grad

    def v_func(u,v):
        grad = (1-u) -v
        return v + dt*grad

    u1 = np.zeros((img_size,img_size))
    v1 = np.zeros((img_size,img_size))

    u1 = np.random.random(u1.shape)+thresh
    u1[u1>0.5]=1
    u1[u1<=0.5]=0
    v1 = 1 - u1

    v1 = np.ones((img_size,img_size))*0.5

    Ku = np.zeros((img_size,img_size))
    Kv = np.zeros((img_size,img_size))

    Ku[0,0]=1+4*du/dx/dx*dt
    Ku[1,0]=-du/dx/dx*dt
    Ku[-1,0]=-du/dx/dx*dt
    Ku[0,1]=-du/dx/dx*dt
    Ku[0,-1]=-du/dx/dx*dt

    Kv[0,0]=1+4*dv/dx/dx*dt
    Kv[1,0]=-dv/dx/dx*dt
    Kv[-1,0]=-dv/dx/dx*dt
    Kv[0,1]=-dv/dx/dx*dt
    Kv[0,-1]=-dv/dx/dx*dt

    Kufi=1/np.fft.fft2(Ku)
    Kvfi=1/np.fft.fft2(Kv)

    for t in range(3):
        for s in range(2000):
            u2 = np.fft.ifft2(np.fft.fft2(u_func(u1,v1))*Kufi)
            v2 = np.fft.ifft2(np.fft.fft2(v_func(u1,v1))*Kvfi)

            u1 = np.fft.ifft2(np.fft.fft2(u_func(u2,v2))*Kufi)
            v1 = np.fft.ifft2(np.fft.fft2(v_func(u2,v2))*Kvfi)

        h = (u1/u1.max())*255
        h = h.astype(np.uint8)
        return h
```

```
h = phase_field(alpha,du,dv,thresh)
plt.imshow(h)
plt.show()
```
