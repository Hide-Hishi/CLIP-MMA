```
import numpy as np
import matplotlib.pyplot as plt
```

```
def rotation(r_x,r_y,theta,theta_rand=False):
    if theta_rand:
        #0.1倍でプラマイ6度が1σのブレ
        rand_list = np.random.randn(len(r_x))*theta*0.2
        theta = rand_list + theta
    else:
        pass
    rotated_x = np.cos(theta)*r_x - np.sin(theta)*r_y
    rotated_y = np.sin(theta)*r_x + np.cos(theta)*r_y
    return rotated_x, rotated_y

def bifurcatedVector(p,theta,p1_x,p1_y,p2_x,p2_y,d,theta_rand = False):
    r_x = p2_x - p1_x
    r_y = p2_y - p1_y
    r1_x,r1_y = rotation(r_x,r_y,theta,theta_rand = theta_rand)
    r2_x,r2_y = rotation(r_x,r_y,-1*theta,theta_rand = theta_rand)
    r1_x *= p
    r1_y *= p
    r2_x *= p
    r2_y *= p

    ARRAY1 = np.array([p2_x, p2_y, p2_x + r1_x, p2_y + r1_y, d+1])
    ARRAY2 = np.array([p2_x, p2_y, p2_x + r2_x, p2_y + r2_y, d+1])
    ARRAY = np.concatenate([ARRAY1,ARRAY2],axis = 1)
    return ARRAY.T

def L_system(n,p = 0.8,theta = 0.7,thickness_p = 0.7,init_thickness = 5,theta_rand = False):
    segmentList =np.array([[0,0,0,1,0]])
    resultList = segmentList[0]
    initial_condition = 0
    #最初の幹を描画
    plt.plot([segmentList[0,0],segmentList[0,1]],[segmentList[0,2],segmentList[0,3]],'k-',lw=init_thickness)

    #枝分かれ回数
    n = 9
    nextL = segmentList
    for degree in range(n):
        nextL = bifurcatedVector(p,theta,nextL[:,0],nextL[:,1],nextL[:,2],nextL[:,3],nextL[:,4],theta_rand = theta_rand)
        segmentList = np.concatenate([segmentList,nextL],axis=0)
        #ここまでで分岐点の計算は終了
        start_x = nextL[:,0]
        start_y = nextL[:,1]
        end_x = nextL[:,2]
        end_y = nextL[:,3]
        thickness = init_thickness*(thickness_p**nextL[0,4])
        #第n層目の枝を描画
        plt.plot([start_x,end_x],[start_y,end_y],'k-',lw=thickness)

    plt.gca().set_aspect('equal')
    plt.axis("off")
    plt.show()

    return segmentList
```

```
#相似比・角度・枝の太さ
p = 0.7
theta =  0.7
#枝が次の分岐で何倍の太さになるか
thickness_p = 0.7
#最初の幹の太さ
init_thickness = 5
```

```
L_system(6,p,theta,thickness_p,init_thickness,theta_rand = True)
```
